```python
import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Initialisation
ticker = yf.Ticker("TSLA")  # Tesla
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)

# Initialisation d'un set pour suivre les UUID déjà ajoutés
seen_uuids = set()
all_news = []

# Exemple de boucle corrigée
current_date = start_date
while current_date < end_date:
    next_date = current_date + timedelta(days=30)  # Intervalle
    news = ticker.news
    for item in news:
        if item["uuid"] not in seen_uuids:
            # Vérifiez que la date est dans l'intervalle
            publish_time = datetime.fromtimestamp(item["providerPublishTime"])
            if current_date <= publish_time < next_date:
                all_news.append({
                    "title": item["title"],
                    "providerPublishTime": publish_time
                })
                seen_uuids.add(item["uuid"])  # Ajoutez l'UUID au set
        print(f"Current date: {current_date}")
        print(f"Next date: {next_date}")
        print(f"Number of articles added: {len(all_news)}")

    current_date = next_date

news = ticker.news
print(news)

# Transformer en DataFrame
tesla_news = pd.DataFrame(all_news)
tesla_news = pd.DataFrame(all_news).drop_duplicates(subset='title')

#tesla_news['date'] = tesla_news['providerPublishTime'].dt.date
print(tesla_news.head())
```

Tentative avec une boucle pour contourner la limite d'un mois, on se rend compte que c'est seulement le dernier mois qui est accessible
